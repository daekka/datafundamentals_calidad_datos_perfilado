```python
import pandas as pd
import matplotlib as mat
import seaborn as sea
```

```python
import pandas as pd

# Cargar el archivo CSV df_combinado.csv
df_trabajo = pd.read_csv('..\\output\\df_combinado.csv')
print(df_trabajo.head())
```

```python
# Crear figura y ejes
fig, ax = mat.pyplot.subplots(figsize=(12, 6))

# Crear gráfica de barras
sea.barplot(x='Tipo_Equipo', y='Costo_Mantenimiento', data=df_trabajo, estimator=sum, ci=None, ax=ax)
ax.set_title('Distribución del Costo de Mantenimiento por Tipo de Equipo')
ax.set_xlabel('Tipo de Equipo')
ax.set_ylabel('Costo de Mantenimiento (suma)')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Mostrar gráfica
mat.pyplot.show()
```

```python
# Crear figura y ejes para temperatura vs horas operativas
fig, ax = mat.pyplot.subplots(figsize=(12, 6))
sea.scatterplot(x='Temperatura_C', y='Horas_Operativas', hue='Tipo_Equipo', data=df_trabajo, ax=ax)
ax.set_title('Relación entre Temperatura y Horas Operativas')
ax.set_xlabel('Temperatura (°C)')
ax.set_ylabel('Horas Operativas')

# Mostrar gráfica
mat.pyplot.show()

# Crear figura y ejes para vibración vs horas operativas
fig, ax = mat.pyplot.subplots(figsize=(12, 6))
sea.scatterplot(x='Vibracion_mm_s', y='Horas_Operativas', hue='Tipo_Equipo', data=df_trabajo, ax=ax)
ax.set_title('Relación entre Vibración y Horas Operativas')
ax.set_xlabel('Vibración (mm/s)')
ax.set_ylabel('Horas Operativas')

# Mostrar gráfica
mat.pyplot.show()
```

```python
# Crear figura y ejes para horas operativas
fig, ax = mat.pyplot.subplots(figsize=(12, 6))
sea.boxplot(x='Tipo_Equipo', y='Horas_Operativas', data=df_trabajo, ax=ax)
ax.set_title('Diagrama de Caja de Horas Operativas por Tipo de Equipo')
ax.set_xlabel('Tipo de Equipo')
ax.set_ylabel('Horas Operativas')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Mostrar gráfica
mat.pyplot.show()

# Crear figura y ejes para costos de mantenimiento
fig, ax = mat.pyplot.subplots(figsize=(12, 6))
sea.boxplot(x='Tipo_Equipo', y='Costo_Mantenimiento', data=df_trabajo, ax=ax)
ax.set_title('Diagrama de Caja de Costos de Mantenimiento por Tipo de Equipo')
ax.set_xlabel('Tipo de Equipo')
ax.set_ylabel('Costo de Mantenimiento')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Mostrar gráfica
mat.pyplot.show()
```

```python
# Crear figura y ejes
fig, ax = mat.pyplot.subplots(figsize=(12, 6))

# Crear gráfica de barras
sea.countplot(x='Tipo_Equipo', hue='Tipo_Mantenimiento', data=df_trabajo, ax=ax)
ax.set_title('Frecuencia de Mantenimiento por Tipo de Equipo y Tipo de Mantenimiento')
ax.set_xlabel('Tipo de Equipo')
ax.set_ylabel('Frecuencia de Mantenimiento')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.legend(title='Tipo de Mantenimiento')

# Mostrar gráfica
mat.pyplot.show()
```
